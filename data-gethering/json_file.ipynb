```python
from tabulate import tabulate
chunk_iter = pd.read_sql(query, engine, chunksize=1)

for i, chunk in enumerate(chunk_iter, start=1):
    print(f"👉 Processing chunk #{i}")
    print(tabulate(chunk.values, headers=list(chunk.columns), tablefmt='psql'))
```
