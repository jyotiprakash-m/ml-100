```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
```

```python
# 1️⃣ Countplot (frequency of categories)
plt.figure(figsize=(6,4))
sns.countplot(x='Survived', data=df)
plt.title("Count of Survived")
plt.show()
```

```python
# 2️⃣ Pie Chart (percentage of categories)
df['Survived'].value_counts().plot.pie(autopct='%1.1f%%', colors=['lightblue','lightgreen'])
plt.title("Survived - Pie Chart")
plt.ylabel("")  # Remove y-label
plt.show()
```

```python
# 3️⃣ Histogram (distribution of a numerical column)
plt.figure(figsize=(6,4))
plt.hist(df['Fare'].dropna(), bins=30, color='skyblue', edgecolor='black')
plt.title("Age Distribution (Histogram)")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
# 4️⃣ Distplot / KDE (probability density)
plt.figure(figsize=(6,4))
sns.histplot(df['Age'].dropna(), kde=True, color='purple')
plt.title("Age Distribution with KDE")
plt.xlabel("Age")
plt.show()
```

```python
# 5️⃣ Boxplot (IQR, outliers)
plt.figure(figsize=(6,4))
sns.boxplot(y='Age', data=df)
plt.title("Boxplot of Age")
plt.show()
```
